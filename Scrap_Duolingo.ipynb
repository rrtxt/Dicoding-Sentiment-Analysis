### Data Scrapping

```python
from google_play_scraper import Sort, reviews_all
```

com.nianticlabs.monsterhunter

```python
scrapped_review = reviews_all(
    "com.duolingo",
    lang='en',
    country='us',
    sort=Sort.MOST_RELEVANT,
    count=500
)
```

```python
import pandas as pd

df = pd.DataFrame(scrapped_review)
df.rename(columns={"content" : "review"}, inplace=True)
df
```

```python
df.to_csv("./dataset/mhnow_review_en.csv", index=False)
df.head()
```
